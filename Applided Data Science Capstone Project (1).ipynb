```python
import pandas as pd # library for data analsysis
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M" 
data=pd.read_html(url, header=0) # Website scraping
data_new = data[0]
data_new=data_new[data_new.Borough != 'Not assigned'] # Drop rows with not assigned values in Borough column
data_new = data_new.groupby(['Postal Code','Borough'], sort = False).agg(lambda x: ','.join(x)) # Combining neighborhoods with the same postal index into one row - no longer needed
data_new.reset_index(level=['Postal Code','Borough'], inplace=True)
data_new.shape



```
